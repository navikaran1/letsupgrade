Q6. Python program to find the largest element among three Numbers ?

num1 = 10

num3 = 15

```python
num1 = 10
num2 = 50
num3 = 14

if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print("The largest number is", largest)
```
